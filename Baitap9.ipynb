```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path =  "../ThiGiacMayTinh/images/06.jpg"
img = Image.open(image_path)
plt.imshow(img)
```

```python
img_gray = img.convert('L')
plt.imshow(img_gray, cmap='gray')
```

```python
img_array = np.array(img_gray)
print(img_array)
```

```python
plt.figure()
plt.hist(img_array.flatten(), bins=255)
plt.title("Biểu đồ Histogram của ảnh")
plt.xlabel("Giá trị của pixel")
plt.ylabel("Số lượng")
plt.show()
```

```python
plt.figure()
plt.gray()
plt.contour(img_gray, origin="image")
plt.show()
```
