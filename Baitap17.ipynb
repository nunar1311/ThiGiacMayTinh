```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

my_path = "../ThiGiacMayTinh/images/01.jpeg"
img = Image.open(my_path)

gray_img = img.convert('L')

img_array = np.array(gray_img)

gradient_x = ndimage.sobel(img_array, axis=0, mode='constant')

fig, axs = plt.subplots(2, 1, figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.imshow(gradient_x, cmap='gray')
plt.title('Gradient Image')
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

my_path = "../ThiGiacMayTinh/images/03.jpeg"
img = Image.open(my_path)

gray_img = img.convert('L')

img_array = np.array(gray_img)

gradient_x = ndimage.convolve(img_array, [[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
gradient_y = ndimage.convolve(img_array, [[-1, -2, -1], [0, 0, 0], [1, 2, 1]])

gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)

fig, axs = plt.subplots(2, 1, figsize=(10, 5))

plt.subplot(2, 1, 1)
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')

plt.subplot(2, 1, 2)
plt.imshow(gradient_magnitude, cmap='gray')
plt.title('Gradient Image')
```
