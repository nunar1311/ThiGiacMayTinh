```python
import numpy as np
from PIL import Image

def average_images(image_list):
    total_array = np.array(Image.open(image_list[0]), 'f')
    count=1

    for image_path in image_list[1:]:
        try:
            image_array = np.array(Image.open(image_path), 'f')
            total_array += image_array
            count=count+1
        except:
            print("Skip ", image_path);

    average_array = total_array / count

    average_image = Image.fromarray(average_array.astype('uint8'))

    return average_image

img_1 = "../ThiGiacMayTinh/images/03.jpeg"
img_2 = "../ThiGiacMayTinh/images/04.jpeg"
img_list = [img_1, img_2]
result_img = average_images(img_list)
display(result_img)

img_1 = "../ThiGiacMayTinh/images/05.jpeg"
img_2 = "../ThiGiacMayTinh/images/06.jpg"

img_list = [img_1, img_2]
result_img = average_images(img_list)
display(result_img)
```
