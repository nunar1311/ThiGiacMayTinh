```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 16, 40, 60, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [30, 15, 8, 9, 50]

plt.plot(x, y, 'ro-', label='Dữ liệu')
plt.plot(x1, y1, marker='o', color='b', linestyle='-', label='Dữ liệu 1')
plt.legend()
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [35, 10, 5, 32, 48]

plt.bar(categories, values, color='g', alpha=0.5)
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

```python
my_labels = ['A', 'B', 'C', 'D']
sizes = [55, 30, 15, 10]
my_colors = ['red', 'green', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%')
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [15, 20, 30, 40, 34]

plt.scatter(x, y, c='r', marker='o', label='Dữ liệu')
plt.title('Biểu đồ phân tán')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend()
plt.show()
```

```python
data = [15, 18, 22, 30, 35, 40, 50, 53, 65]

plt.boxplot(data)
plt.title('Biểu đồ hộp ví dụ')
plt.ylabel('Giá trị')
plt.show()
```

```python
data = [1, 10, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data)
plt.title('Biểu đồ violin')
plt.ylabel('Giá trị')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "Python is an amazing language for data visualization and word clouds."

wordcloud = WordCloud(width=800, height=400).generate(text)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Biểu đồ đám mây từ văn bản')
plt.show()
```

```python
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 4, 2, 8, 5]

N = len(categories)
values += values[:1]

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

plt.polar(angles, values, 'o-', linewidth=2)
plt.fill(angles, values, 'b', alpha=0.1)
plt.title('Biểu đồ radar')
plt.show()
```
