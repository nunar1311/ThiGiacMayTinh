```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

image_path =  "../ThiGiacMayTinh/images/02.jpeg"
img = Image.open(image_path)

img_array = np.array(img.convert("L"))

inveted_img = 255 - img_array
squared_img = inveted_img ** 2
clamped_img = np.clip(squared_img, 100, 200)

fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(img_array, cmap="gray")
axs[0, 0].set_title("Ảnh gốc")

axs[0, 1].imshow(inveted_img, cmap="gray")
axs[0, 1].set_title("Ảnh biến đổi")

axs[1, 0].imshow(squared_img, cmap="gray")
axs[1, 0].set_title("Ảnh biến đổi")

axs[1, 1].imshow(clamped_img, cmap="gray")
axs[1, 1].set_title("Ảnh biến đổi")

plt.show()
```
