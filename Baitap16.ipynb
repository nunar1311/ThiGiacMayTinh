```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage

my_path = "../ThiGiacMayTinh/images/05.jpeg"
img = Image.open(my_path)

gray_img = img.convert('L')

img_arr = np.array(gray_img)

blurred_image_a = ndimage.gaussian_filter(img_arr, sigma=1, mode="constant")
blurred_image_b = ndimage.gaussian_filter(img_arr, sigma=3, mode="constant")
blurred_image_c = ndimage.gaussian_filter(img_arr, sigma=5, mode="constant")

fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_img, cmap='gray')
plt.title('nh xm gc')

plt.subplot(2,2,2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('sigma=1')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('sigma=3')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('sigma=5')

```
