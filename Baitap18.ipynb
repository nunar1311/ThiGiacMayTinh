```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

my_path = "../ThiGiacMayTinh/images/07.png"
img = Image.open(my_path)

gray_img = img.convert('L')

img_arr = np.array(gray_img)

binary_img = 1 * (img_arr > 200)

opened_img = ndimage.binary_opening(binary_img, structure=np.ones((3, 3)))

labeled_img, num_features = ndimage.label(opened_img)

fig, axs = plt.subplots(3, 1, figsize=(30, 30))

plt.subplot(4, 1, 1)
plt.imshow(gray_img, cmap='gray')
plt.title('Original Image')

plt.subplot(4, 1, 2)
plt.imshow(binary_img, cmap='gray')
plt.title('Binary Image')

plt.subplot(4, 1, 3)
plt.imshow(opened_img, cmap='gray')
plt.title('Opened Image')

plt.subplot(4, 1, 4)
plt.imshow(labeled_img, cmap='gray')
plt.title(f'Labeled ({num_features}) features')
```
