```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def histogram_equalization(image, nbr_bins=256):
    if image.mode != "L":
      image = image.convert("L")

      img_array = np.array(image)

      histogram, bins = np.histogram(img_array, bins=nbr_bins, range=(0, 256), density=True)

      cdf = histogram.cumsum()
      cdf = 255 * cdf / cdf[-1]

      image_equalized = np.interp(img_array, bins[:-1], cdf)

      equalized_img = Image.fromarray(image_equalized.astype("uint8"))

      return equalized_img
```

```python
my_path = "../ThiGiacMayTinh/images/06.jpg"

img = Image.open(my_path)

equalized_img = histogram_equalization(img)

plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(img.convert("L"), cmap='gray')
plt.title('Hnh nh gc')

plt.subplot(2, 2, 2)
plt.hist(np.array(img.convert("L")).flatten(), bins=128)
plt.title('Biu  Histogram - Hnh nh gc')

plt.subplot(2, 2, 3)
plt.imshow(equalized_img, cmap='gray')
plt.title('Hnh nh  cn bng')

plt.subplot(2, 2, 4)
plt.hist(np.array(equalized_img).flatten(), bins=128)
plt.title('Biu  Histogram - Hnh nh  cn bng')

plt.show()
```
